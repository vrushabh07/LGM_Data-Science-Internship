# Task 1 : Image to Pencil Sketch with Python 

Import Libraries

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

Read image 

```python
image = cv2.imread("oreo.jpg")
cv2_imshow(image)
#plt.imshow(image)
#plt.show()
```

Convert the image from original to grayscale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2.waitKey(0)
```

Resized an image to 600*600



```python
resized = cv2.resize(image,(600,600))
resized
```

Inverted the new grayscale image

```python
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
```

Blurr the image by using gaussian function

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

Invert the blurred image.

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```

Comparing the original and sketch photo

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
